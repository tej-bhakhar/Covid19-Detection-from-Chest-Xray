```python
# import all neccesary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import shutil
```

```python
# Create the data for positive samples

FILE_PATH="chestxray/metadata.csv"
IMAGE_PATH="chestxray/images"
```

```python
df=pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
TARGET_DIR="dataset/covid"

if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("covid folder created")
```

```python
cnt=0

for (i,row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        filename=row["filename"]
        imagepath=os.path.join(IMAGE_PATH,filename)
        imagecopy=os.path.join(TARGET_DIR,filename)
        shutil.copy2(imagepath,imagecopy)

        cnt +=1
        
    
    
```

```python
# sampling of images from kaggle

import random
kaggle_file_path="chest_xray/train/NORMAL"
target_normal_dir="dataset/Normal"
```

```python
image_names=os.listdir(kaggle_file_path)
```

```python
random.shuffle(image_names)
```

```python
for i in range(142):
    image_name=image_names[i]
    image_path=os.path.join(kaggle_file_path,image_name)
    target_path=os.path.join(target_normal_dir,image_name)
    
    shutil.copy2(image_path,target_path)
```
